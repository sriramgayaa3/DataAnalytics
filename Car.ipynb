```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
sns.lineplot(x="Year", y="Price", data=a, estimator="mean", ci=None, marker="o")
plt.title("Average Car Price Over Years")
plt.xlabel("Year")
plt.ylabel("Average Price")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(x="Fuel_Type", y="Price", data=a, estimator="mean", palette="viridis")
plt.title("Average Price by Fuel Type")
plt.xlabel("Fuel Type")
plt.ylabel("Average Price")
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.scatterplot(x="Mileage", y="Price", data=a, alpha=0.5, color="b")
plt.title("Car Price vs Mileage")
plt.xlabel("Mileage")
plt.ylabel("Price")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x="Transmission", y="Price", data=a, palette="coolwarm")
plt.title("Car Price Distribution by Transmission Type")
plt.xlabel("Transmission Type")
plt.ylabel("Price")
plt.show()
```

```python
fuel_counts = a["Fuel_Type"].value_counts()
plt.figure(figsize=(7, 7))
plt.pie(fuel_counts, labels=fuel_counts.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette("Set2"))
plt.title("Fuel Type Distribution")
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(a["Price"], bins=30, kde=True, color="g")
plt.title("Distribution of Car Prices")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(a["Mileage"], bins=30, kde=True, color="purple")
plt.title("Distribution of Car Mileage")
plt.xlabel("Mileage")
plt.ylabel("Frequency")
plt.show()
```

```python
numeric_a = a.select_dtypes(include=['number'])
corr_matrix = numeric_a.corr()
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Matrix")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x="Engine_Size", y="Price", data=a, palette="Set1")
plt.title("Car Price Distribution by Engine Size")
plt.xlabel("Engine Size")
plt.ylabel("Price")
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.countplot(x="Year", data=a, palette="deep")
plt.xticks(rotation=45)
plt.title("Number of Cars by Year")
plt.xlabel("Year")
plt.ylabel("Count")
plt.show()
```
